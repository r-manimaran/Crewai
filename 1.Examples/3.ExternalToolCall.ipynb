```python
# .venv\Scripts\python -m pip install crewai[tools]     
from crewai import Agent, Task, Crew, LLM
from crewai_tools import SerperDevTool
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
import os

load_dotenv(dotenv_path="C:/Maran/Study/Git/Crewai/1.Examples/.env", override=True)
print(os.getenv("OPENAI_API_KEY")) 
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
os.environ["SERPER_API_KEY"] = os.getenv("SERPER_API_KEY")
print(os.getenv("GOOGLE_API_KEY"))
print(os.getenv("SERPER_API_KEY"))
```

```python
llm = LLM(model="gemini/gemini-2.5-flash", temperature=0.0)
search_tool = SerperDevTool()
```

```python
researcher_agent = Agent(
    goal = "Search the web and provide accurate and up-to-date information.",
    role ="Web researcher",
    backstory = "You are a web researcher agent who is an expert in finding accurate and up-to-date information on the internet. You have access to a powerful search tool that allows you to quickly gather information from various sources. Your primary goal is to find the name of the current chief minister of Tamil Nadu and provide a concise answer in one line.",
    tools = [search_tool],
    llm = llm,
    verbose = True
)
```

```python
task = Task(
    description ="Who is the cheif-minister of Tamil Nadu in 2026? Answer in one line.",
    expected_output="A short summary of the current chief minister of Tamil Nadu.",
    agent=researcher_agent
)

crew = Crew(Agents=[researcher_agent], tasks=[task], verbose=True)
result = crew.kickoff()
print("Final Result:", result)
```
