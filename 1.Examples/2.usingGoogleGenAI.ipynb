```python
from crewai import Agent, Task, Crew, LLM
from dotenv import load_dotenv
from langchain_google_genai import GoogleGenerativeAI
import os
# Load environment variables from .env file
load_dotenv(dotenv_path="c:/Maran/Study/Git/Crewai/.env", override=True)
print(os.getenv("GOOGLE_API_KEY"))  # verify it loaded
```

```python
gemini_llm = LLM(model="gemini/gemini-2.5-flash", temperature=0)
```

```python
# create the researcher agent
researcher_agent = Agent(
    llm=gemini_llm,
    role=".NET Interview question researcher",
    goal ="Research and compile a list of 10 common .NET interview questions and their answers.",
    backstory="You are an experienced .NET developer with a deep understanding of common interview questions and answers. Your task is to research and compile a list of 10 common .NET interview questions along with their answers to help candidates prepare for their interviews.",
    allow_delegation=False, #prevent the agent from delegating tasks to other agents
    verbose=True
)

# create a writer agent
writer_agent = Agent(
    llm=gemini_llm,
    role=".NET Interview question writer",
    goal ="Write a comprehensive guide on .NET interview questions and answers based on the research provided by the researcher agent.",
    backstory="You are a skilled technical writer with expertise in .NET development. Your task is to write a comprehensive guide on .NET interview questions and answers based on the research provided by the researcher agent. The guide should be clear, concise, and informative to help candidates prepare effectively for their interviews.",
    allow_delegation=False, #prevent the agent from delegating tasks to other agents
    verbose=True
)
```

```python
researcher_task = Task(
    description="Research and compile a list of 10 common .NET interview questions and their answers.",
    agent=researcher_agent,
    expected_output="A list of 10 common .NET interview questions and their answers."
)

writer_task = Task(
    description="Write a comprehensive guide on .NET interview questions and answers based on the research provided by the researcher agent.",
    agent=writer_agent,
    expected_output="A comprehensive guide on .NET interview questions and answers."
)
```

```python
crew = Crew(agents=[researcher_agent, writer_agent], tasks=[researcher_task, writer_task])
result =crew.kickoff()
print(result)
```
