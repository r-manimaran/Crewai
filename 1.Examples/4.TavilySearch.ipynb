```python
from crewai import Agent, Task, Crew, LLM, Process
from langchain_community.tools.tavily_search import TavilySearchResults
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
import os
```

```python
# Load environment variables from .env file
load_dotenv(dotenv_path="c:/Maran/Study/Git/Crewai/.env", override=True)
print(os.getenv("OPENAI_API_KEY"))  # verify it loaded
```

```python
llm= LLM(model="openai/gpt-4o-mini", temperature=0.7)
```

```python
engineer = Agent(
    role="Senior Software Engineer",
    llm=llm,
    goal="Explain the programming concepts with clear code examples",
    backstory=(
        "You are a professional engineer with ample years of experience. \n"
        "You give accurate and concise answers to programming questions, with code examples when necessary. \n"
        "You handle all programming languages, coding and algorithm questions with equal expertise. \n"
    ),
    allow_delegation=False
)
```

```python
writer = Agent(
    llm=llm,
    role="Technical Writer",
    goal="Write a clear and concise article on the given programming topic.",
    backstory=(
        "You are a skilled technical writer with a knack for simplifying complex programming concepts. \n"
        "You produce well-structured articles that are easy to understand, even for beginners. \n"
        "Your writing is engaging and informative, making use of examples and analogies to clarify difficult topics. \n"
    ),
    allow_delegation=False
)
```

```python
task1 = Task(
    description=("Explain the concept of Entity Framework.\n"
    "Include a code example in .NET\n"
    "Make it concise and beginner-friendly."),
    expected_output="A clear and concise explanation with .NET code example.")

task2 = Task(
    description=(
        "Write a story about moon\n"
        "Make it concise and beginner-friendly."
    ),
    expected_output="A beginner-friendly explanation. using simple language"
)

     
```

```python
crew = Crew(agents=[engineer, writer], tasks=[task1],
process=Process.hierarchical,
verbose=True,
manager_llm=llm
)
result = crew.kickoff()
print("Final Result:", result)
```
